# 爬取昵图网图片

```python
import bs4
import requests
import pprint
import os
import threading

os.makedirs('NiTu', exist_ok = True)

def downLoadPic(newurl,headers):
    print("Starting download %s "%newurl)
    response = requests.get(newurl,headers=headers)
    response.raise_for_status()
    response.encoding='utf-8'
    soup = bs4.BeautifulSoup(response.text,'html.parser')
    link_elements = soup.select("div[class='photo'] > a > img")
    num = 1
    for url in link_elements:
        downloadUrl = url.get("picurl")
        print('Downloading %s image:\n %s ...' % (num,downloadUrl))
        res1 = requests.get(downloadUrl,headers=headers)#
        if res1.status_code != 200:
            print("cann't download " + downloadUrl)
            continue
        res1.raise_for_status()
        imageFile = open(os.path.join('NiTu', str(num)+"_"+url.get("alt")+'.jpg'),'wb')# os.path.basename(comUrl))返回comUrl末尾那段
        for chunk in res1.iter_content(100000):
            imageFile.write(chunk)
        imageFile.close()
        print(' Succeed to Download %s image:\n %s ...' % (num,downloadUrl))
        num += 1
    print("You have downloaded %s pic"%str(num))
        
def getPageNum(url,headers):#获取页面数
    response = requests.get(url,headers=headers)
    response.raise_for_status()
    response.encoding='utf-8'
    soup = bs4.BeautifulSoup(response.text,'html.parser')
    res = soup.select("span[class='pagerV2'] > span")
    return len(res)
        
if __name__ == "__main__":
    url = "http://www.huitu.com/topic-detail/3243.html" #昵图网荷花图片
    headers = {"User-Agent":"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36"}
    pagenum = getPageNum(url,headers) + 1
    downloadThreads = []
    for i in range(1,pagenum):
        if i != 1:
            newurl = url[:-5]+"-"+str(i)+url[-5:]
        else:
            newurl = url
        #nitupic = getNiTuPic(newurl,headers)
        
        downloadThread = threading.Thread(target=downLoadPic,args=(newurl,headers))
        downloadThreads.append(downloadThread)
        downloadThread.start()
        #nitupic.downLoadPic()
    for downloadThread in downloadThreads:
        downloadThread.join()
    print('Done:\n')
    #nitupic = getNiTuPic(url,headers)
    #nitupic.downLoadPic()
```
